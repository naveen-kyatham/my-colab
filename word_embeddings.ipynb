```python
import nltk
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from gensim.models import Word2Vec

# Download the necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Define the input text
text = "A quick brown fox jumped over the street"
# Tokenize the text into sentences
sentences = nltk.sent_tokenize(text)
# Tokenize each sentence into words
tokenized_sentences = [word_tokenize(sentence) for sentence in sentences]
# Remove stopwords
stop_words = set(stopwords.words('english'))
#print(stop_words)
tokenized_sentences = [[token for token in sentence if token not in stop_words] for sentence in tokenized_sentences]


# Create a Word2Vec model
model = Word2Vec(tokenized_sentences, vector_size=100, window=5, min_count=1, workers=8)


# Tokenize the text
#tokens = word_tokenize(text)


# Create a Word2Vec model
#model = Word2Vec(tokens, vector_size=100, window=5, min_count=1, workers=8)

# Print the word embeddings
#print(tokenized_sentences)
for word in tokenized_sentences:
  embedding = model.wv[word]
 # print(f"{word}: {embedding}")
#type(embedding)
row_names = ['A', 'quick', 'brown', 'fox', 'jumped', 'street']
#need to be dynamic
#=[sentence[0] for sentence in tokenized_sentences]
df = pd.DataFrame(embedding)
df.insert(0, 'word', row_names)
df
```
